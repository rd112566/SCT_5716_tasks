```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd


india_pop_df = pd.DataFrame({
    'Year': ['2019', '2020', '2021', '2022', '2023'],
    'Population': [1.38903e9, 1.402618e9, 1.414204e9, 1.425423e9, 1.438070e9]
})

plt.figure(figsize=(8, 5))
sns.barplot(x='Year', y='Population', data=india_pop_df, palette='Blues_d')

for index, row in india_pop_df.iterrows():
    plt.text(index, row['Population'] + 1e7, f"{row['Population'] / 1e9:.2f}B", ha='center')

plt.title("India's Total Population (2019â€“2023)")
plt.ylabel("Population")
plt.xlabel("Year")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
